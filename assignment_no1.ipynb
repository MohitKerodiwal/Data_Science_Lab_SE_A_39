```python
# 1. WAP to compute factorial of a given number
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage:
# num_fact = 5
# print(f"The factorial of {num_fact} is {factorial(num_fact)}")
```

```python
# 2. WAP to display table for a given number
def display_multiplication_table(num):
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")

# Example usage:
# num_table = 7
# display_multiplication_table(num_table)
```

```python
# 3. WAP to identify the given number is odd or even
def check_odd_even(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Example usage:
# num_oe = 10
# print(f"{num_oe} is {check_odd_even(num_oe)}")
# num_oe = 7
# print(f"{num_oe} is {check_odd_even(num_oe)}")
```

```python
# 4. WAP to identify the entered number is palindrome or not.
def is_palindrome_number(num):
    # Convert number to string to easily reverse it
    s_num = str(num)
    if s_num == s_num[::-1]:
        return True
    else:
        return False

# Example usage:
# num_pal = 121
# print(f"{num_pal} is a palindrome: {is_palindrome_number(num_pal)}")
# num_pal = 123
# print(f"{num_pal} is a palindrome: {is_palindrome_number(num_pal)}")
```

```python
# 5. Develop an Expert System that provides simple decision-making.
# This is a very basic example of an expert system, usually they are more complex
# and involve rule engines or machine learning models.

def simple_expert_system(symptom):
    symptom = symptom.lower()
    if "fever" in symptom and "cough" in symptom:
        return "You might have a cold or flu. Rest and drink fluids."
    elif "headache" in symptom and "nausea" in symptom:
        return "Consider resting in a dark room. Consult a doctor if symptoms persist."
    elif "hungry" in symptom:
        return "You should eat something!"
    else:
        return "I'm sorry, I cannot provide advice for that. Please consult a professional."

# Example usage:
# print(simple_expert_system("I have a fever and cough"))
# print(simple_expert_system("feeling hungry"))
# print(simple_expert_system("my car won't start"))
```
