```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate a normal distribution
# Using numpy.random.normal(mean, standard_deviation, number_of_samples)
mean = 0
std_dev = 1
num_samples = 1000
normal_distribution_data = np.random.normal(mean, std_dev, num_samples)

# Plot its histogram
plt.hist(normal_distribution_data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Histogram of a Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Assuming 'normal_distribution_data' from the previous cell is available

# 2. Compute mean, median, variance, and standard deviation

# Mean
data_mean = np.mean(normal_distribution_data)
print(f"Mean: {data_mean:.4f}")

# Median
data_median = np.median(normal_distribution_data)
print(f"Median: {data_median:.4f}")

# Variance
data_variance = np.var(normal_distribution_data)
print(f"Variance: {data_variance:.4f}")

# Standard Deviation
data_std_dev = np.std(normal_distribution_data)
print(f"Standard Deviation: {data_std_dev:.4f}")
```
